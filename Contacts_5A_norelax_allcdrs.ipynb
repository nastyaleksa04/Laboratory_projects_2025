**Разницы энергий TCRen:**

delta=structure_2 (B2702) - structure (B2705)


```python
structure_pairs = [
    ('7n2o', '7n2o_2'),
    ('7n2q', '7n2q_2'),
    ('8cx4', '8cx4_2')
]

structure_energy_map = {
    struct: all_energy_df[all_energy_df['Structure'] == struct].set_index('Region')
    for struct in all_energy_df['Structure'].unique()
}

delta_energy_rows = []

for struct1, struct2 in structure_pairs:
    df1 = structure_energy_map.get(struct1)
    df2 = structure_energy_map.get(struct2)
    
    if df1 is None or df2 is None:
        continue

    merged = df1[['Energy']].rename(columns={'Energy': 'Energy_1'}).join(
        df2[['Energy']].rename(columns={'Energy': 'Energy_2'}),
        how='inner'
    )
#delta=structure_2 (B2702) - structure (B2705)

    merged['Delta_Energy'] = merged['Energy_2'] - merged['Energy_1']
    merged['Structure_1'] = struct1
    merged['Structure_2'] = struct2
    merged = merged.reset_index()[['Region', 'Structure_1', 'Structure_2', 'Delta_Energy']]
    
    delta_energy_rows.append(merged)
delta_energy_df = pd.concat(delta_energy_rows, ignore_index=True)
```

**Разница множеств контактов**

```python
def extract_contact_set(df):
    df_filtered = df[df['residue.aa.to'].notna()]
    return set(zip(
        df_filtered['Region'],
        df_filtered['Chain'],
        df_filtered['Residue Number'],
        df_filtered['residue.aa.from'],
        df_filtered['residue.aa.to'],
        df_filtered['Contacted Peptide Number']
    ))
```

```python
structure_pairs = [
    ('7n2o', '7n2o_2'),
    ('7n2q', '7n2q_2'),
    ('8cx4', '8cx4_2')
]


all_results = []

for struct1, struct2 in structure_pairs:
    df1 = globals()[f'df_{struct1}']
    df2 = globals()[f'df_{struct2}']

    set1 = extract_contact_set(df1)
    set2 = extract_contact_set(df2)

    union_contacts = set1 | set2

    for contact in union_contacts:
        all_results.append({
            'Region': contact[0],
            'Chain': contact[1],
            'Residue Number': contact[2],
            'residue.aa.from': contact[3],
            'residue.aa.to': contact[4],
            'Contacted Peptide Number': contact[5],
            'In_structure_1': contact in set1,
            'In_structure_2': contact in set2,
            'Structure_1_name': struct1,
            'Structure_2_name': struct2
        })

result_df = pd.DataFrame(all_results)
```

```python
mj_k = pd.read_csv('MJ_Keskin_potentials.csv')

df_keskin = mj_k[mj_k['potential'] == 'Keskin'].reset_index(drop=True)
df_mj = mj_k[mj_k['potential'] == 'MJ'].reset_index(drop=True)

structure_list = ['7n2o', '7n2o_2', '7n2q', '7n2q_2', '8cx4', '8cx4_2']  
result_by_structure_new = {}

for structure in structure_list:
    df = globals()[f'df_{structure}']
    df.columns = ['Region', 'Chain', 'Residue Number', 'residue.aa.from',
                  'Contact Count', 'Min Distance (Å)', 'residue.aa.to',
                  'Contacted Peptide Number']
    
    merged_df_new = pd.merge(df, df_mj, on=['residue.aa.from', 'residue.aa.to'], how='inner')
    merged_df_new['MJ_potential'] = merged_df_new['value'] * merged_df_new['Contact Count']

    energy_sum_new = merged_df_new.groupby("Region")["MJ_potential"].sum().reset_index()

    result_by_structure_new[structure] = energy_sum_new

all_energy_df_new = pd.concat(
    [
        df.assign(Structure=structure)
        for structure, df in result_by_structure_new.items()
    ],
    ignore_index=True
)

all_energy_df_new = all_energy_df_new[['Structure', 'Region', 'MJ_potential']]


grouped = all_energy_df_new.groupby(['Structure', 'Region'], as_index=False)['MJ_potential'].sum()

hla_grouped_new = grouped[grouped['Region'].isin(['HLA alpha1', 'HLA alpha2'])]

hla_summed_new = hla_grouped_new.groupby('Structure', as_index=False)['MJ_potential'].sum()
hla_summed_new['Region'] = 'HLA'

non_hla_new = grouped[~grouped['Region'].isin(['HLA alpha1', 'HLA alpha2'])]

all_energy_df_new = pd.concat([non_hla_new, hla_summed_new], ignore_index=True)

all_energy_df_new = all_energy_df_new[['Structure', 'Region', 'MJ_potential']]
print(all_energy_df_new)

cdr_regions = ['CDR1 alpha', 'CDR2 alpha', 'CDR3 alpha', 'CDR1 beta', 'CDR2 beta', 'CDR3 beta']
cdr_df_new = all_energy_df_new[all_energy_df_new['Region'].isin(cdr_regions)]

cdr_sums_new = cdr_df_new.groupby('Structure', as_index=False)['MJ_potential'].sum()
cdr_sums_new.rename(columns={'MJ_potential': 'CDR_sum'}, inplace=True)

print("Суммарный потенциал MJ по CDR:")
print(cdr_sums_new)

structure_list = ['7n2o', '7n2o_2', '7n2q', '7n2q_2', '8cx4', '8cx4_2']  
result_by_structure = {}

for structure in structure_list:
    df = globals()[f'df_hla_tcr_{structure}']
    df.columns = ['CDR Region', 'CDR Chain', 'CDR Residue Number', 'residue.aa.from',
                  'HLA Region', 'HLA Chain', 'HLA Residue Number', 'residue.aa.to',
                  'Distance (Å)']
    
    merged_df = pd.merge(df, df_mj, on=['residue.aa.from', 'residue.aa.to'], how='inner')
    merged_df['MJ_potential'] = merged_df['value'] * 1

    energy_sum = merged_df.groupby("CDR Region")["MJ_potential"].sum().reset_index()

    result_by_structure[structure] = energy_sum
    
    
    
all_energy_df = pd.concat(
    [
        df.assign(Structure=structure)
        for structure, df in result_by_structure.items()
    ],
    ignore_index=True
)

all_energy_df = all_energy_df[['Structure', 'CDR Region', 'MJ_potential']]


cdr_regions = ['CDR1 alpha', 'CDR2 alpha', 'CDR3 alpha', 'CDR1 beta', 'CDR2 beta', 'CDR3 beta']
cdr_df = all_energy_df[all_energy_df['CDR Region'].isin(cdr_regions)]

cdr_sums = cdr_df.groupby('Structure', as_index=False)['MJ_potential'].sum()
cdr_sums.rename(columns={'MJ_potential': 'CDR_sum'}, inplace=True)

print("MJ по HLA-CDR:")
print(cdr_sums)

df_final = all_energy_df_new[all_energy_df_new['Region'] == 'HLA']
cdr_sums_new['MJ_potential'] = cdr_sums_new['CDR_sum']
cdr_sums_new['Region'] = 'CDR'
cdr_sums_new = cdr_sums_new.drop(['CDR_sum'], axis=1)
cdr_sums['MJ_potential'] = cdr_sums['CDR_sum']
cdr_sums['Region'] = 'TCR-HLA'
cdr_sums = cdr_sums.drop(['CDR_sum'], axis=1)
df_final = pd.concat([df_final, cdr_sums_new], ignore_index=True)
df_final = pd.concat([df_final, cdr_sums], ignore_index=True)

def normalize_structure(name):
    return name.replace('_2', '')

df_final['Pair'] = df_final['Structure'].apply(normalize_structure)

idx_min = df_final.groupby(['Pair', 'Region'])['MJ_potential'].idxmin()
df_min = df_final.loc[idx_min].copy()

def assign_b270(name):
    return 'B2702' if name.endswith('_2') else 'B2705'

df_min['HLA_type'] = df_min['Structure'].apply(assign_b270)

result = df_min[['Pair', 'Region', 'MJ_potential', 'HLA_type']].reset_index(drop=True)
print('Минимальная энергия:')
print(result)
```
