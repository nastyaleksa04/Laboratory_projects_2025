```python
import pandas as pd
```

```python
norelax_5A = pd.read_csv('/home/aalexandrova/structures_norelax/comparing_methods/norelax_5A.csv')
norelax_getcontacts = pd.read_csv('/home/aalexandrova/structures_norelax/comparing_methods/norelax_getcontatcs.csv')
rosetta_5A = pd.read_csv('/home/aalexandrova/structures_norelax/comparing_methods/rosetta_5A.csv')
rosetta_getcontacts = pd.read_csv('/home/aalexandrova/structures_norelax/comparing_methods/rosetta_getcontacts.csv')
tcrmodel_5A = pd.read_csv('/home/aalexandrova/structures_norelax/comparing_methods/tcrmodel_5A.csv')
tcrmodel_getcontacts = pd.read_csv('/home/aalexandrova/structures_norelax/comparing_methods/tcrmodel_getcontacts.csv')
```

```python
norelax_5A
```

```python
def annotate_df(df, method, is_5a, is_getcontacts):
    df = df.copy()
    df['Methods'] = method
    df['5A'] = is_5a
    df['Getcontacts'] = is_getcontacts
    return df

# Аннотирование каждого датафрейма
dfs = [
    annotate_df(norelax_5A, 'norelax', True, False),
    annotate_df(norelax_getcontacts, 'norelax', False, True),
    annotate_df(rosetta_5A, 'rosetta', True, False),
    annotate_df(rosetta_getcontacts, 'rosetta', False, True),
    annotate_df(tcrmodel_5A, 'tcrmodel', True, False),
    annotate_df(tcrmodel_getcontacts, 'tcrmodel', False, True)
]

# Объединение всех в один датафрейм
combined_df = pd.concat(dfs, ignore_index=True)

# # Расставим столбцы в нужном порядке
combined_df = combined_df[['Methods', '5A', 'Getcontacts', 'Energy', 'HLA_type', 'Pair', 'Region']]

# # (опционально) Сортировка, например:
combined_df.sort_values(by=['Region', 'Energy'], inplace=True)

# print(combined_df.head())
```
